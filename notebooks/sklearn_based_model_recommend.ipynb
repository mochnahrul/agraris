```python
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split

from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
file_name = "dataset/Crop_recommendation.csv"
df = pd.read_csv(file_name)
df
```

```python
labelEncoder = LabelEncoder()

labelEncoder.fit(df["label"])
labelEncoder_key_value = dict(zip(labelEncoder.classes_, labelEncoder.transform(labelEncoder.classes_)))

pd.DataFrame(labelEncoder_key_value.items(), columns=['Original Label', 'Encoded'])
```

```python
df["label"] = labelEncoder.transform(df["label"])

df
```

```python
minmaxScaler = MinMaxScaler()
minmaxScaler.fit(df.iloc[:, 0:-1])
df.iloc[:, 0:-1] = minmaxScaler.transform(df.iloc[:, 0:-1])
```

```python
df
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = svm.SVC()
model_2 = KNeighborsClassifier()
model_3 = RandomForestClassifier(n_estimators=3)

model.fit(X_train, y_train)
model_2.fit(X_train, y_train)
model_3.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred2 = model_2.predict(X_test)
y_pred3 = model_3.predict(X_test)
```

```python
print("Akurasi model SVM: {}%".format(round(accuracy_score(y_test, y_pred)*100, 2)))
print("Akurasi model KNN: {}%".format(round(accuracy_score(y_test, y_pred2)*100, 2)))
print("Akurasi model RF: {}%".format(round(accuracy_score(y_test, y_pred3)*100, 2)))
```

```python
print("Confusion Matrix:\n\n{}".format(confusion_matrix(y_test, y_pred3)))
```
